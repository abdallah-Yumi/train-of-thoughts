## common indicators of a movie succsess

```python
ass_grouping = data [['original_title', 'cast','director' ,'production_companies','budget','revenue', 'genres', 'release_date']]
ass_grouping
```

## as shown above the movie industry shares a common indicators of success yet there's thousands of movies 

### but what are the criterias associated with success, budget\revenue, ratings or popularity

## well let's check the relevance between them and determine what the sucsess factor is 

### starting with cost/revenue analysis to the top 50 

```python
cost_rev = ass_grouping.copy()
cost_rev.loc[:, 'profit'] = cost_rev['revenue'] - cost_rev['budget']
cost_rev_most = cost_rev.sort_values(by='profit', ascending=False)
cost_rev_most.head(50)
```

### these are the most profitable moives ever 

### but what are the factors that makes it succsessful? 

### let's investigate the objective factors that effects it
#### company or cast

```python
comp_rev = cost_rev_most.copy()
comp_rev = comp_rev.merge(data[['original_title', 'production_companies']], on='original_title')



comp_rev.head(50)
```

```python
compnum = comp_rev.copy()
compnum['company_count'] = compnum.groupby('production_companies_x').cumcount() + 1
compnum_sorted = compnum.sort_values(by='company_count', ascending=False)


compnum.head(50)
```

## although the data is inaccurate theres a big inication that production of corprate companies are involved in succsession 

### but how many of them were involved in the top 50 moives

```python
comp_copy = cost_rev.copy()
comp_copy = comp_copy.merge(data[['original_title', 'production_companies']], on='original_title')
comp_copy = comp_copy.sort_values(by='revenue', ascending=False)
comp_copy_filtered = comp_copy[comp_copy['revenue'] > 960000000]


disney_count = (comp_copy_filtered['production_companies'].str.contains('Disney', case=False)).sum()
uni_count = (comp_copy_filtered['production_companies'].str.contains('universal', case=False)).sum()
marvel_count = (comp_copy_filtered['production_companies'].str.contains('marvel', case=False)).sum()
warner_count = (comp_copy_filtered['production_companies'].str.contains('warner', case=False)).sum()
warner_count = (comp_copy_filtered['production_companies'].str.contains('warner', case=False)).sum()
para_count = (comp_copy_filtered['production_companies'].str.contains('paramount', case=False)).sum()
col_count = (comp_copy_filtered['production_companies'].str.contains('columbia', case=False)).sum()


print(f'The word "Disney" appears {disney_count} times in the production companies.')
print(f'The word "universal" appears {uni_count} times in the production companies.')
print(f'The word "marvel" appears {marvel_count} times in the production companies.')
print(f'The word "warner bros" appears {warner_count} times in the production companies.')
print(f'The word "paramount pictures" appears {para_count} times in the production companies.')
print(f'The word "columbia pictures" appears {col_count} times in the production companies.')
comp_copy_filtered.head(50)
```
